```python
# Importing libraries

from kafka import KafkaConsumer
from time import sleep
from json import dumps,loads
import json
from s3fs import S3FileSystem
```

```python
# Creating a KafkaConsumer instance

consumer = KafkaConsumer(
    
    # Specify the Kafka topic to consume messages from
    'demo_testing2', 
    
    # Specifying the Kafka broker addresses (bootstrap servers, replace the ip and port number below)                         
                         bootstrap_servers=[#'ip:port number'],
                             
    # Defining a custom value deserializer using JSON deserialization
                        value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
# Create an instance of S3FileSystem

s3 = S3FileSystem()
```

```python
# Iterate over messages received by the Kafka consumer along with their index (count)
for count, i in enumerate(consumer):
    
    # Use the S3FileSystem instance to open an S3 object for writing(Replace the bucketname/projectname below for desired filename)
    with s3.open("s3://bucketname/projectname_{}.json".format(count), 'w') as file:
        
        # Serialize and write the value of the Kafka message (i.value) to the S3 object using JSON
        json.dump(i.value, file)
```
